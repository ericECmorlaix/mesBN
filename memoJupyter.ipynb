## Pour disposer d'un environnement interactif Jupyter :
### Local :
- La solution la plus pratique est d'installer une distribution Anaconda : https://www.anaconda.com/distribution/
 
### En ligne :

- Chez Microsoft Azure Notebooks : https://notebooks.azure.com/
    > *Le travail de Tony HIRST permet d'explorer : https://notebooks.azure.com/OUsefulInfo/projects/gettingstarted*

- Chez Google Colaboratory : https://colab.research.google.com
    > *Google a particip  l'origine de la cration du projet Jupyter puis a poursuivi son dveloppement indpendemment pour des besoins internes avant que de le rendre public en l'ajoutant  sa suite d'outils bureautique. C'est diffrent mais compatible et permet le partage...*

- Pour ouvrir de faon interactive mais temporaire un dpot GitHub ou GitLab dans un environnement Jupyter : https://mybinder.org/.
    > ***Attention**, il faut ajouter un [fichier de configuration](https://mybinder.readthedocs.io/en/latest/using.html#preparing-a-repository-for-binder) soit en [pip](https://github.com/binder-examples/requirements) soit en [conda](https://github.com/binder-examples/conda) pour indiquer  la machine virtuelle les paquets (Noyaux, Modules, Extensions)  installer.*

## Extensions et/ou Noyaux complmentaires :
>*Certaines distributions viennent avec des noyaux (Kernels) et des extensions dj intgres.*

- Diaporama RISE : https://github.com/damianavila/RISE
    > [Tutoriel de Microsoft Azure Notebooks pour RISE](https://docs.microsoft.com/fr-fr/azure/notebooks/present-jupyter-notebooks-slideshow)

- [nbconvert](https://nbconvert.readthedocs.io) pour convertir les Notebook du format `.ipynb` vers `.pdf`, `.tex`,...

- [metakernel](https://github.com/Calysto/metakernel) pour inclure des codes Blockly, Processing, et bien plus encore...: [documentation](http://calysto.github.io/metakernel/), [exemples](http://nbviewer.jupyter.org/github/Calysto/metakernel/tree/master/examples/) : [Python Tutor](https://jupyter.brynmawr.edu/services/public/dblank/Examples/Tutor%20Magic%20in%20IPython.ipynb), ...

- [nbtutor](https://github.com/lgpage/nbtutor) pour visualiser l'excution d'un code Python ligne par ligne dans un notebook tel qu'avec [Python Tutor](https://github.com/lgpage/nbtutor), lire [l'article de Tony Hirst](https://blog.ouseful.info/2017/01/27/python-code-stepper-debugger-tutor-for-jupyter-notebooks-nbtutor/)  ce sujet.

- [nbautoeval](https://github.com/parmentelat/nbautoeval) une solution pour crer des exercices d'autovaluation en python dans un Jupyter notebook.

- [Jupyter-PHP](https://litipk.github.io/Jupyter-PHP-Installer/) un noyau pour php.

- la collection [Unofficial Jupyter Notebook Extensions](https://github.com/ipython-contrib/jupyter_contrib_nbextensions) qui intgre notamment :
    - le gestionnaire d'extensions [jupyter_nbextensions_configurator](https://github.com/Jupyter-contrib/jupyter_nbextensions_configurator) ;
    - [Variable Inspector](https://jupyter-contrib-nbextensions.readthedocs.io/en/latest/nbextensions/varInspector/README.html) pour afficher l'tat des variables dans une fentre flottante ;
    - [gist-it](https://medium.com/deena-does-data-science/jupyter-notebook-extensions-e6b57d004e8e) pour publier un jupyter notebook sur GitHub en un clic.
    
> *Ces extensions ne fonctionnent pas dans JupyterLab*

## Jupyterlab : 
> L'environnement Jupyter du futur !
>
> Pour basculer de l'environnement Jupyter NoteBook classique  l'environnement Lab, il suffit de remplacer dans l'URL le `tree` par un `lab` et inversement...
>
>C est une nouvelle gnration de linterface web utilisateur, plus volue, qui intgre les applications de Notebook et de Terminal classiques et adjoint une console Ipython et de nouvelles fonctionnalits graphiques pour la gestion des fichiers de lespace personnel et laffichage interactif simultan sur de multiples fentres.
>
> De plus, cet environnement autorise lintgration dapplications en extensions comme un diteur de diagrammes ou encore un diteur/compilateur LaTeX
